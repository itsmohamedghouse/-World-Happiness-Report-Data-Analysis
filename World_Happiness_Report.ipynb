### **Linear Regression (univariate)**

Linear regression is a common Statistical Data Analysis technique. It is used to determine the extent to which there is a linear relationship between a dependent variable and one or more independent variables.

#### **Selecting "GDP per capita" as it is highly correlated with happiness score**


```
x=data['GDP per capita'].values
y=data['Score'].values
```

```
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```
plt.plot(x,y,'r.')
plt.xlabel('GDP per capita')
```

```
def h(a,x,b):
  return (a*x) + b
```

```
def error(a,x,b,y):
  m = len(y)
  e = 0
  for i in range(m):
    e += np.power(h(a,x[i],b)-y[i], 2)
  
  return e/(2*m)
```

```
def step_gradient(a,x,b,y,learning_rate):
  grad_a = 0
  grad_b = 0
  m = len(y)
  for i in range(m):
    grad_a += (1/m) * (h(a,x[i],b) - y[i]) * x[i]
    grad_b += (1/m) * (h(a,x[i],b) - y[i])

  a = a - (grad_a * learning_rate)
  b = b - (grad_b * learning_rate)

  return a,b
```

```
def descend(init_a,x,init_b,y,learning_rate,iter):
  a = init_a
  b = init_b

  for i in range(iter):
    if i%1000==0:
      print(f"Error: {error(a,x,b,y)} | a: {a} | b: {b}")
    a,b = step_gradient(a,x,b,y,learning_rate)

  return a, b
```

```
a = 0
b = 0 
learning_rate = 0.001
```

```
plt.plot(xtrain, ytrain, 'r.',
         xtrain, h(a, xtrain,b), 'b')
```

```
a, b = descend(a,xtrain,b,ytrain,learning_rate,10000)
```

```
plt.plot(xtrain, ytrain, 'r.',
         xtrain, h(a, xtrain,b), 'b')
```

```
np.sqrt(error(a, xtrain,b,ytrain))
```

```
np.sqrt(error(a, xtrain,b,ytrain))
```
